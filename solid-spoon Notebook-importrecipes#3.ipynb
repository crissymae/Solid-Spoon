```python
def create_dummy(df, column, split):
    f_set=set()
    for feature in df[column]:
        feature_list = feature.split(split)
        for f in feature_list:
            f_set.add(f)
    f_lst=list(f_set)
    for item in f_lst:
        df[item]=df[column].apply(lambda x: 1 if item in x else 0)
    new_df=df.drop(columns=column)
    return new_df
```
