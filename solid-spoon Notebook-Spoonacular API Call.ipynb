# API Call for Spoonacular Data


```python
# Import standard packages
import requests
import config
import json
import pandas as pd
```

```python
APIKey = config.SpoonacularAPIKey
```

```python
resp = requests.get (f'https://api.spoonacular.com/recipes/complexSearch?apiKey={APIKey}&offset=0&number=1&addRecipeInformation=True&addRecipeNutrition=True&fillIngredients=True')
```

```python
resp = requests.get (f'https://api.spoonacular.com/recipes/complexSearch?apiKey={APIKey}&offset=100&number=1&addRecipeInformation=True&addRecipeNutrition=True')
```

```python
resp.status_code
```

```python
#Having a look at the json file to view data
resp.json()
```

```python
#Creating a function that uses requests to pull down multiple recipes at time - seems to be a max of 1000 recipes
compresults=[]
for i in range(50):
    resp = requests.get (f'https://api.spoonacular.com/recipes/complexSearch?apiKey={APIKey}&offset={i*100}&number=100&addRecipeInformation=True&addRecipeNutrition=True&fillIngredients=True').json()
    try:
        compresults.extend(resp['results'])
    except:
        print("No results found for {}".format(i))
    
```

```python
#Taking a peek at the data in df format
df = pd.DataFrame(compresults)
df
```

```python
#Viewing one row of data in order to view all column names
display(df.iloc[1])
```
